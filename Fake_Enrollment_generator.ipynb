### Create functions that will take the list of teacher names, shirt sizes, schools, and instruments and map them onto random integers, thus randomly assigning each student one of each attribute from the available options.

```python
# create shirt size function to generate random shirt size from choices
def shirt_assigner():
    # create shirt sizes dictionary to map sizes onto random ints from numpy
    sizes = {1: "XS",
         2: "S",
         3: "M",
         4: "L",
         5: "XL"
        }
    # generate random int
    return sizes[np.random.randint(1,6)]
```

```python
# create function to map teacher names onto randit
def teacher_assigner():
   


    # create dictionary private teachers, us numpy random to assign an int, then map that int to a private teacher
    teachers = {0: "",
            1: "Bucket",
            2: "Alexander-Bucket",
            3: "Tibbs",
            4: "Gunter",
            5: "Cakelake",
            6: "McFly",
            7: "del Rio",
            8: "Wrestler",
            9: "Restler",
            10: "Rosenctrantz",
            11: "Gildenstern",
            12: "Rodgers",
            13: "Opie",
            14: "Silver",
            15: "Kitten",
            16: "Bull"
           }
    return teachers[np.random.randint(0,17)]
```

```python
# create function to map instruments onto randint, include argument for primary or secondary inst, as primary inst can't be blank
def instrument_assigner(primary = True):  
    # create dictionary instruments, us numpy random to assign an int, then map that int to an instrument
    instruments = {0: "",
                  1: "0 Soprano",
                  2: "1 Alto",
                  3: "2 Tenor",
                  4: "3 Baritone"}
    # if primary = True, then only include 1-4 since primary can't be blank, else include 0-4
    if primary:
        inst_num = np.random.randint(1,5)
    else:
        inst_num = np.random.randint(0,5)
    return instruments[inst_num]
    
```

```python
# create schools function for mapping
# include argument for grade level in order to correctly assign a middle or high school
def school_assigner(grade):
    # Create middle schools dictionary to map random ints onto schools
    middle_schools = {1: "Bee Cave MS",
                      2: "Canyon Ridge MS",
                      3: "Cedar Park MS",
                      4: "Cele MS",
                      5: "Chisolm Trail",
                      6: "Covington MS",
                      7: "Danielson MS",
                      8: "Deerpark MS",
                      9: "Four Points MS",
                      10: "Grisham MS",
                      11: "Henry MS",
                      12: "Lake Travis MS",
                      13: "Leander MS",
                      14: "Manor New Tech MS",
                      15: "Running Brushy MS",
                      16: "Small MS",
                      17: "Stiles MS",
                      18: "Walsh MS",
                      19: "Wiley MS"
                     }
    # create high_schools dict
    high_schools = {1: "Anderson HS",
                    2: "Austin HS",
                    3: "Cedar Park HS",
                    4: "James Bowie HS",
                    5: "Lake Travis HS",
                    6: "Round Rock HS",
                    7: "Vandegrift HS",
                    8: "Vista Ridge HS",
                    9: "Weiss HS"
                   }
    # assign a school based on grade level
    if grade > 8:
        return high_schools[np.random.randint(1,10)]
    else: 
        return middle_schools[np.random.randint(1,20)]
```
